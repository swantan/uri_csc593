1) Load either your dataset, `brfss`, or `nhis` and create at least three derived fields from the data there. Ideally, these will be fields you'll need for your final project.

```python
fwff = '../data/brfss/LLCP2018.ASC'

        
names= ['state', 'imonth', 'iday', 
        'iyear', 'dispcode','genhlth', 
        'physhlth',
        'menthlth', 'poorhlth', 'hlthpln1',
        'persdoc2', 'medcost', 'checkup1',
        'WEIGHT2', 'HEIGHT3']
cols = [
    (1, 3),
    (18, 20),
    (20, 22),
    (22, 27),
    (31, 35),
    (89, 90),
    (90, 92),
    (92, 94),
    (94, 96),
    (96, 97),
    (97, 98),
    (98, 99),
    (99, 100),
    (176, 180),
    (180, 184)
]
types= {
    'WEIGHT2': str, 
    'HEIGHT3': str,
}
brfss = pd.read_fwf(fwff + ' ', 
                    names=names,
                    colspecs=cols,
                    dtype=types)
```

```python
print(brfss.shape)
#Drop any rows without weight
brfss.dropna(subset=['WEIGHT2', 'HEIGHT3'], inplace = True)
print(brfss.shape)
```

```python
np.unique(brfss.dispcode)
```

```python
#to create a derived field that describe dispcode -- final disposition
brfss['dispcode_desc'] = np.where(brfss.dispcode == 1100, 'completed interview', 'partial complete')
brfss.head()
```

```python
np.unique(brfss.physhlth)
```

```python
# mh = 1 if you were sick more than 7 days, 0 otherwise:
brfss['mh'] = np.where((brfss['menthlth'] > 30) | (brfss['menthlth'] <= 7) | (brfss['menthlth'].isnull()), 0, 1)
brfss.head()
```

```python
## 7 means "Don't know" / "Not sure"; want to group vague answer like don't know, refused and missing values
#and just consider informative answer
brfss.loc[brfss.checkup1 == 8, 'checkup1'] = 5
bins = [0, 1, 2, 3, 4, 5]
brfss['checkupcat'] = pd.cut(brfss.checkup1, bins, 
                         labels=['less than 12 mths', 'less than 2 yrs', 
                                 'less than 5 yrs', 'more than 5 yrs', 'never'])
brfss.head()
```

```python
np.unique(brfss.checkupcat)
```
